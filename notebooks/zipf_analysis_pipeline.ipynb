```python
import sys
import os

# Ajouter le dossier src au chemin d'import
sys.path.append(os.path.abspath("../src"))

from zipf.preprocessing import charger_texte, nettoyer_texte, tokenizer_texte
from zipf.frequency import compter_mots
from zipf.plotting import tracer_zipf

texte = charger_texte("C:/code/projets_perso/zipf-law-project/data/raw/Corpus_Victor_Hugo.txt")
texte = nettoyer_texte(texte)
tokens = tokenizer_texte(texte)
compteur, stopwords = compter_mots(tokens)
mots_freq = compteur.most_common(100)
tracer_zipf(mots_freq, debut=1)
```
